# Homogeneous Coordinates(동차 좌표)

일반적으로 점 A = (1, 2, 3) 이렇게만 표시하면 벡터인지 포인트인지 알 방법이 없습니다.

따라서 w 라는 변수를 추가하여 0이면 벡터, 1이면 포인트로 정의하는 방법을 도입했습니다.

A = (a, b, c)

A = (a, b, c, w)

A = (1, 2, 3, 1) 이면 점 A, A = (1, 2, 3, 0) 이면 벡터 A

이렇게 네 가지 성분으로 표현하는 방식을 homogeneous coordinates 라고 정의합니다.

homogeneous coordinates로 표현된 점과 벡터를 조작하는 것은 이 좌표들의 기하학적 해석과 호환됩니다.

예를 들어,

두 점의 좌표를 빼면 벡터로, 두 벡터를 더하면 벡터로 표현되는 것처럼 점과 벡터를 더하면 점이 생성되고, 벡터에 스칼라를 곱하면 벡터가 생성됩니다.

### 0이나 1이 아닌 w값을 가진 좌표가 의미하는 것? 

1이상의 w값은 큰 의미를 갖진 않습니다. w값 자체의 의미보단 w로 각 좌표의 값을 나눴을 때의 그 비율이 중요합니다. 좌표가 달라도 그 비율이 같다면 같은 좌표입니다.

즉, (1, 2, 3, 1) 과 (2, 4, 6, 2), (-3, -6, -9, -3)은 전부 동일한 좌표입니다.
