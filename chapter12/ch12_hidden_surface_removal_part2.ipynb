## Using 1/Z instead of Z

https://mgun.tistory.com/2176

앞선 결과는 이전보다 나아진 것을 확인할 수 있다. 다만 약간 잘못된 부분이 있다.

vertice들로부터 가져온 Z값들은 주어진 데이터로부터 가져왔기 때문에 정확하지만 선형 보간을 통해 구해낸 나머지 픽셀들의 Z값들은 정확하지 않다. 지금 당장은 차이가 크게 보이지 않지만 나중에 언제든지 문제가 될 수 있는 부분이다.

예시를 통해 어떤 부분이 잘못되었는지 알아보자. 두 점 $A(-1, 0, 2)$과 $B(1,0,10)$으로 이루어진 선분이 있다. 선분 위에는 두 점 사이의 중간점인 $M(0,0,6)$이 존재한다. 구체적으로 $M_z = (A_z + B_z)/2 = 6$을 통해 중간점을 구한다는 것을 우리는 알고있다.

![image](https://user-images.githubusercontent.com/74246766/150086930-60dba81f-39e6-4cfe-bba3-d844af860e8f.png)

</br>
</br>
</br>
이 점들을 $Z=d=1$로 프로젝션 해보자. 우리는 원근 투영 공식을 이용하여 $A'_x = A_x / A_z = -1/2 = -0.5$을 얻을 수 있다. 마찬가지로 $B'_x = 0.1$, $M'_x = 0$을 얻을 수 있다.

![image](https://user-images.githubusercontent.com/74246766/150087242-3ce446e9-8d53-42cb-8031-f5dc31dee2c2.png)

</br>
</br>
</br>
$A'B'$는 viewport 상에서 수평한 선분이다. 우리는 $A_z$와 $B_z$의 값을 알고있다. 이제 선형 보간을 이용하여 $M_z$값을 구하려 할 때 어떤 일이 일어나는지 알아보자. 암시적인 선형 함수는 아래 그림과 같다.

![image](https://user-images.githubusercontent.com/74246766/150088264-0c580edd-0412-4d5b-b1c3-626a7c226e46.png)

</br>
</br>
</br>
함수의 기울기는 일정하므로 다음과 같은 식을 작성할 수 있다.

$$\frac{M_z - A_z}{M'_x - A'_x} = \frac{B_z - A_z}{B'_x - A'_x}$$

$M_z$에 대해 정리해보자.

$$M_z = A_z + (M'_x - A'_x)(\frac{B_z - A_z}{B'_x - A'_x})$$

각 점에 값들을 넣어보면 다음과 같다.

$$M_z = 2 + (0 - (-0.5))(\frac{10 - 2}{0.1 - (-0.5)}) = 2 + (0.5)(\frac{8}{0.6}) = 8.666$$

위 계산에 따르면 $M_z$는 8.666의 값을 갖는다. 그러나 우리는 실제 값이 6이라는 사실을 알고 있다. 무엇이 잘못된 것일까.

우리의 실수는 우리가 선형 보간을 사용할 때 만들었던 숨겨진 가정에 있다. 바로 우리가 보간하려는 함수는 선형 함수라는 것이다. 이 경우에는 그렇지 않다.

만약 $Z = f(x',y')$가 $x',y'$의 선형 함수일 때 우리는 $Z = Ax' + By' + C$와 같이 표현할 수 있다. 이런 종류의 함수는 두 점 간의 값의 차이는 점 사이의 차이에는 의존하지만 점 자체에는 의존하지 않는다는 특성이 있다.

$$f(x' + \Delta x, y' + \Delta  y) - f(x', y') = [A(x' + \Delta  x) + B(y' + \Delta  y) + C] - [A \cdot x' + B \cdot y' + C] $$
</br>
$$= A(x' + \Delta  x - x') + B(y' + \Delta  y - y') + C - C $$
</br>
$$= A\Delta  x + B\Delta  y$$

</br>
</br>
즉, 좌표가 다르더라도 좌표의 차이가 동일하다면 Z의 차이도 동일하다.

더 공식적으로(more formally), 선분을 포함하는 평면의 방정식은 다음과 같다.

$$Ax + By + Cz + D = 0$$

한편 projection equation은 다음과 같다.

$$x' = \frac{x \cdot d}{z}$$

$$y' = \frac{x \cdot d}{z}$$

위 공식으로부터 x와 y을 얻으면

$$x = \frac{z \cdot x'}{d}$$

$$y = \frac{z \cdot y'}{d}$$

만약 평면 방정식에 위 식을 대입하면 다음과 같다.

$$\frac{Ax'z + By'z}{d} + Cz + D = 0$$

이를 d로 나누고 z에 대해 풀면

$$Ax'z + By'z + dCz + dD = 0$$

$$(Ax' + By' + dC)z + dD = 0$$

$$z = \frac{-dD}{Ax' + By' + dC}$$

위 식은 x'와 y'에 대한 선형함수가 당연히 아니다. 그리고 이것이 선형 보간이 잘못된 z값을 도출한 이유이다.

그러나 z 대신에 1/z값을 이용해보자.

$$1/z = \frac{Ax' + By' + dC}{-dD}$$

위 함수는 x'과 y'에 대한 선형 함수이다. 즉 1/z에 대한 선형 보간이 올바른 값을 도출할 것을 알 수 있다.

이번엔 $M_z$를 1/z를 통해 구해보자.

$$\frac{M_\frac1z - A_\frac1z}{M'_x - A'_x} = \frac{B_\frac1z - A_\frac1z}{B'_x - A'_x}$$

$$M_\frac1z = A_\frac1z + (M'_x - A'_x)(\frac{B_\frac1z - A_\frac1z}{B'_x - A'_x})$$

$$M_\frac1z = \frac12 + (0 - (-0.5))(\frac{\frac1{10} - \frac12}{0.1 - (-0.5)}) = 0.166666$$

$$M_z = \frac1{M_\frac1z} = \frac1{0.166666} = 6$$

이 값은 이전에 기하학을 기반으로 한 $M_z$값과 일치한다는 점에서 정확하다.

이것이 우리가 z 대신 1/z을 사용해야하는 이유이다. 추가적으로 반영되어야 하는 차이는 모든 버터의 시작은 0으로 초기화되어야 한다는 것과(z가 무한대가 되면 1/z는 0이 되어야 한다.) 비교 부등호가 뒤집혀야 한다는 것(z값이 클수록 1/z값은 작다.)이다.
