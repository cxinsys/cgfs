## Normal Mapping

13장에서 우리는 표면의 법선 벡터가 물체의 외관에 얼마나 큰 영향을 미치는지 알 수 있었다. 예를 들어 법선 벡터의 올바른 선택이 깎여있는 객체가 올바르게 보이도록 한 것을 알 수 있었다.

![image](https://user-images.githubusercontent.com/74246766/151169384-dac6e186-97d7-4d65-aff6-bd1a7bd7a20e.png)
![image](https://user-images.githubusercontent.com/74246766/151169404-3bb16b49-9e28-41df-aac9-0b8c2ec6554f.png)

이것은 법선 벡터를 어떻게 선택하느냐에 따라 빛이 물체 표면과 상호작용하는 방식이 바뀌고 이에 따라 우리 뇌가 물체의 모양을 추측하는 방식이 바뀌기 때문이다. 아쉽게도 표면을 매끄럽게 보이게 하기 위해 법선 벡터를 보간했던 우리의 방식 외에 더 발전할 만한 방식은 없다.

14장에서 우리는 texture mapping을 통해 물체의 표면에 그림을 그리는 작업을 진행했다. 이 기술을 사용하여 표면을 더 세밀하게 제어할 수 있었지만 텍스쳐 메핑은 삼각형의 구조를 바꾸진 못한다. 즉, 여전히 평평하다.

$Normal \ mapping$은 두 아이디어를 결합한 것이다. 법선 벡터를 사용하여 빛의 상호작용을 바꾸어 표면의 형태를 바꾼다. 속성 메핑을 이용하여 삼각형의 각기 다른 부분에 다른 속성 값을 추가할 수 있다. 두 아이디어를 결합함으로 인해 normal mapping은 우리에게 픽셀 단위에서 표면의 법선 벡터를 정의할 수 있도록 해 준다.

이를 위해서 우리는 각 삼각형에 normal map을 연결한다. 노멀 맵은 텍스처 맵과 비슷하지만 각 요소들이 색깔이 아닌 법선 벡터로 이루어져 있다. 랜더링 할 때, 이전의 Phong shading에서 그랬던 것처럼 법선 벡터들을 보간하지 않고, 대신 노말 맵에서 우리가 랜더링하려는 픽셀의 법선 벡터 값을 가져오면 된다. 텍스처 매핑인 특정 위치에서의 색 값을 가져왔던 것처럼 이 벡터를 픽셀에서의 빛 계산에 사용한다.

아래 그림은 노말 맵이 적용되었을 때와 그렇지 않을 때를 나타낸 그림이다.

![image](https://user-images.githubusercontent.com/74246766/151171042-e8fd876a-93e7-4766-925a-f9b4a525acf4.png)

여기에서 염두해야할 몇 가지 고려사항이 있다. 첫 번째는, 노멀 맵의 벡터의 방향은 맵이 적용되는 삼각형의 표면을 기준으로 한다. 이러한 좌표계를 접선 공간이라고 한다. 랜더링 시에 카메라 공간에 있는 삼각형의 법선 벡터는 최종적인 조명 방정식을 위해 노멀 맵의 법선 벡터에 따라 수정된다. 이러한 특성은 노말 맵이 씬 안의 물체들의 위치, 회전으로부터 독립되게 해 준다.

두 번째는 노말맵을 텍스처와 같은 형식으로 인코딩한다는 것이다. $X, Y, Z$값을 $R, G, B$에 매칭시킨다. 이것이 노말 맵이 대체적으로 푸르스름한 이유이다. https://raypop.tistory.com/71

![image](https://user-images.githubusercontent.com/74246766/151174150-0d535a84-9b5e-48ff-8c8b-34e19009fc60.png)

노말맵 기술은 표면을 표현하기 위한 복잡성을 크게 개선할 수 있지만 한계가 없는 것은 아니다. 예를 들어 평평한 표면은 계속 평평하게 남아있기 때문에 오브젝트의 실루엣을 변경할 수 없다. 같은 이유로 표면을 너무 극단적인 각도에서 바라보거나 너무 가까이서 보게 되면 노멀맵의 값들이 표면의 크기보다 너무 커버려서 환상이 깨질 수 있다. 이 기술은 피부의 모공, 치장 벽토 벽의 패턴 또는 오렌지 껍질의 불규칙한 모양과 같은 미묘한 세부 사항에 더 적합하다. 이러한 이유로 $bump\ mapping$이라고도 불리운다.
