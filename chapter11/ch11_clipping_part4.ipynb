# Clipping Triangle


sphere - plane 테스트가 개체가 클리핑 평면의 전면에 있는지 또는 완전히 뒤에 있는지 여부를 결정하기에 충분하지 않은 경우 각 삼각형을 해당 평면에 대해 잘라야 한다.\
 평면까지의 부호를 통해 클리핑 평면에 대해 삼각형의 각 꼭짓점을 분류할 수 있다. \
거리가 0이거나 양수이면 정점은 클리핑 평면 앞에 있습니다. 그렇지 않으면 뒤에 있습니다. 그림 11-12는 이 아이디어를 보여준다.

* 세개의 정점이 앞에 있을 경우
* 뒤에 3개의 꼭짓점이 있는 경우
* 하나의 정점이 앞에 있을 경우
* 두개의 정점이 앞에 있을 경우

위에 4개로 평면의 위치를 나눌 수 있다.

* 세 개의 정점이 앞에 있을 경우
전체의 삼각형이 클리핑 평면앞에 있는 경우로 이를 통해 평면에 대해 더 이상 클리핑 할 필요가 없다.


* 뒤에 3개의 꼭짓점이 있는 경우\
전체 삼각형이 클리핑 평면 뒤에 있으므로 버리고 더 이상 클리핑 할 필요가 없다.


* 하나의 정점이 앞에 있을 경우

A가 B'과 C'위에 존재한다면 새로운 AB'C' 삼각형을 추가한다.

* 두개의 꼭짓점이 앞에 있다면.

위 그림처럼 A와 B가 앞에 존재한다면 삼각형 ABC를 지우고 두개의 삼각형을 추가한다. \
ABA' 삼각형과 A'BB' 삼각형을 추가하게 된다. 

# Segment-Plane Intersection

위와 같이 4가지로 분류한 방법으로 삼각형으로 클립하기 위해서는 삼각형의 변과 클리핑 평면의 교차점을 계산해야 한다.

n 평면 벡터와  점 p 에 대한 내적을 한 값 \
$<N, P> + D  = 0 $
\
점 p를 아래와 같이 나타낼 수 있다.


$ P=A+t(B−A) $


$ ⟨ N, A + t(B−A) ⟩ + D = 0 $


$ ⟨ N, A ⟩ + t⟨ N, B−A ⟩ + D = 0 $
