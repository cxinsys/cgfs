## Defining the Cliping Planes

프로젝션 평면인  $Z = d$ 에서 시작한다.    
위 방정식은 시각화하기 편하지만, 우리의 목적에 가장 편하고 일반적인 방정식은 아니다.   

3D 평면의 일반적인 방정식은 $Ax + By + Cz + d =0$ 이다. 
$P$ 가 평면에 있는 경우에 점 $P = (x,y,z)$ 가 방정식을 만족한다는 것을 이야기한다.   

우리는 계수 $(A,B,C)$ 를 $\vec{N}$ 이라 하고 다음과 같이 방정식을 작성할 수 있다. 

$\langle\vec{N},P\rangle + D = 0$

위의 수식에 아무 $k$를 곱해 다음과 같은 방정식을 얻을 수 있다.   

$k\langle\vec{N},P\rangle + kD = 0$      

 
우리는 $k = 1/|\vec{N}|$ 로 정하고 원 방정식에 곱하면 다음과 같은 새로운 방정식을 얻을 수 있다.   

$\langle\vec{N'},P\rangle + D' = 0$      

$\vec{N'}$은 단위 벡터이다.   

따라서 평면은 방정식 $\langle\vec{N},P\rangle + D = 0$ 로 표현될 수 있는데,   
방정식에서 $\vec{N}$ 은 단위 벡터이고, $D$ 는 실수이다. 

$\vec{N}$ 은 평면의 법선이고, $-D$ 는 평면의 원점에서의 *부호가 있는 거리(signed distance)* 이다.   
사실 $\langle\vec{N},P\rangle + D$ 는 $P$에서 평면까지의 부호가 있는 거리이다.    
distance가 0 인 경우는 $P$ 가 평면 위의 점인 특별한 경우이다. 

$\vec{N}$가 평면의 법선벡터이면 $-\vec{N}$ 도 법선 벡터이므로, 우리는 clipping 공간의 안쪽 점을 선택하도록 $\vec{N}$을 선택할 수 있다.   

$Z = d$ 평면일때, 우리는 카메라에 대해 앞쪽 방향을 이야기하는 $(0,0,1)$ 법선벡터를 선택할 수 있다.   
$(0,0,d)$ 는 평면에 속해 있고, 이는 평면의 방정식을 만족시키기 때문에 우리는 D를 해결할 수 있다.   

$\langle\vec{N},P\rangle + D = \langle(0,0,1),(0,0,d)\rangle + D = d + D = 0$   
우리는 $D = -d$ 를 얻을 수 있다. 

우리는 원래의 평면 방정식 $Z = d$에서 $Z - d = 0$ 으로 다시 작성하여 $D = -d$를 즉시 얻을 수 있다.   
하지만, 우리는 나머지 clipping 평면의 방정식을 얻기 위해 위와 같이 일반적인 방법을 적용한다.     

우리는 추가적인 평면은 $D = 0$(원점을 지나기 때문) 임을 알기 때문에, 그들의 법선벡터만 정의하면 된다.   

수학적으로 쉽게 만들기 위해 우리는 FOW(Field of View)를 90도로 정하였다. 이는 평면이 45도 각도임을 이야기한다.   

왼쪽 clipping 평면을 생각해보자.   

법선 벡터는 $(1,0,1)$(45도 각도로 오른쪽 앞 방향) 이다.  
벡터의 길이는 $\sqrt{2}$ 이기 때문에, 우리는 정규화하여 $(\frac{1}{\sqrt{2}},0,\frac{1}{\sqrt{2}})$ 를 얻을 수 있다.   
그러므로 왼쪽 clipping 평면의 방정식은 다음과 같다.   

$\langle\vec{N},P\rangle + D = \langle(\frac{1}{\sqrt{2}},0,\frac{1}{\sqrt{2}}),P\rangle= 0$   

비슷하게, 오른쪽, 위, 아래의 clipping 평면의 법선 벡터는 다음과 같이 정의된다.   
$(\frac{-1}{\sqrt{2}},0,\frac{1}{\sqrt{2}})$ (오른쪽)   
$(0,\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}})$ (아래)   
$(0,\frac{-1}{\sqrt{2}}, \frac{1}{\sqrt{2}})$ (위)   

clipping 공간은 아래의 5 평면으로 정의된다.   

$\langle(0,0,1),P\rangle - d = 0$ (가까운 평면)   

$(\frac{1}{\sqrt{2}},0,\frac{1}{\sqrt{2}})$(왼쪽 평면)

$(\frac{-1}{\sqrt{2}},0,\frac{1}{\sqrt{2}})$ (오른쪽 평면)   

$(0,\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}})$ (아래 평면 )   

$(0,\frac{-1}{\sqrt{2}}, \frac{1}{\sqrt{2}})$ (위 평면)
